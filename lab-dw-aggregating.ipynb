```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv")
df.columns
```

```python
filtered_df = df[(df['Total Claim Amount'] < 1000) & (df['Response'] == 'Yes')]
filtered_df.head()
```

```python
yes_df = df[df['Response'] == 'Yes']
```

```python
summary = yes_df.groupby(['Policy Type', 'Gender'])[
    ['Monthly Premium Auto', 'Customer Lifetime Value', 'Total Claim Amount']
].mean().round(2)

summary
```

```python
state_counts = df['State'].value_counts().reset_index()
state_counts.columns = ['State', 'Customer_Count']
state_counts.head()
```

```python
large_states = state_counts[state_counts['Customer_Count'] > 500]
large_states
```

```python
clv_stats = (
    df.groupby(['Education', 'Gender'])['Customer Lifetime Value']
    .agg(['min', 'median', 'max'])
    .round(2)
    .reset_index()
)

clv_stats
```

```python
df['Effective To Date'] = pd.to_datetime(df['Effective To Date'])
df['Month'] = df['Effective To Date'].dt.month_name()
```

```python
policy_table = pd.pivot_table(
    df,
    values='Customer Lifetime Value',   # any column works, we just need a count
    index='State',
    columns='Month',
    aggfunc='count',
    fill_value=0
)
```

```python
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
policy_table = policy_table.reindex(columns=month_order)
```

```python
policy_table.head()
```

```python
state_month = (
    df.groupby(['State', 'Month'])
    .size()
    .reset_index(name='Policies_Sold')
)
```

```python
top_states = (
    df['State']
    .value_counts()
    .head(3)
    .index
)
```

```python
top_state_data = state_month[state_month['State'].isin(top_states)]
```

```python
top_state_pivot = top_state_data.pivot_table(
    index='State',
    columns='Month',
    values='Policies_Sold',
    fill_value=0
)
```

```python
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
top_state_pivot = top_state_pivot.reindex(columns=month_order)
```
